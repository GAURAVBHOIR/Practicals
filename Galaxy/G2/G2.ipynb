```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix,ConfusionMatrixDisplay
```

```python
# Step 1: Load the data using pandas
data = pd.DataFrame(np.load('/home/gauravbhoir147/Documents/Files/Practicals/Galaxy/G2/galaxies.npy'))

# Step 2: Split the data into features (X) and target (y)
X = data.iloc[:, :-1]  # Features: all columns except the last one
y = data.iloc[:, -1]   # Target: last column

h = [0.2, 0.5, 0.9]
```

```python
for test_size in h:
    # Step 3: Split the data into training and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    # Step 4: Set up a decision tree classifier
    dt_classifier = DecisionTreeClassifier(random_state=42)

    # Step 5: Train the decision tree classifier
    dt_classifier.fit(X_train, y_train)

    # Step 6: Apply the classifier to the test data
    y_pred = dt_classifier.predict(X_test)

    # Step 7: Evaluate accuracy using confusion matrix
    accuracy_dt = accuracy_score(y_test, y_pred)
    print(f"Decision Tree Accuracy with test size {test_size}: {accuracy_dt:.2f}")

    # Plot confusion matrix for decision tree
    cm_dt = confusion_matrix(y_test, y_pred)
    ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
    # Print confusion matrix for decision tree
    print(f"Confusion Matrix - Decision Tree Classifier with test size {test_size}:\n{cm_dt}\n")

    # Step 8: Set up a random forest classifier
    rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

    # Step 9: Train the random forest classifier
    rf_classifier.fit(X_train, y_train)

    # Step 10: Apply the classifier to the test data
    y_pred_rf = rf_classifier.predict(X_test)

    # Step 11: Evaluate accuracy using confusion matrix
    accuracy_rf = accuracy_score(y_test, y_pred_rf)
    print(f"Random Forest Accuracy with test size {test_size}: {accuracy_rf:.2f}")

    # Plot confusion matrix for random forest
    cm_rf = confusion_matrix(y_test, y_pred_rf)
    ConfusionMatrixDisplay.from_predictions(y_test,y_pred_rf)
    # Print confusion matrix for random forest
    print(f"Confusion Matrix - Random Forest Classifier with test size {test_size}:\n{cm_rf}\n")
```
