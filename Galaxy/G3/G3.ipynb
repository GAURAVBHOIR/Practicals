```python
from galpy.potential import (plotRotcurve,MWPotential2014 as mwp14)
import matplotlib.pyplot as plt
import galpy.potential
MWP = mwp14
```

```python
plotRotcurve(MWP,label='Total')
plotRotcurve(MWP[0],label='Bulge',overplot=True)
plotRotcurve(MWP[1],label='Disk',overplot=True)
plotRotcurve(MWP[2],label='Halo',overplot=True)
plt.legend()
plt.show()
```

```python
from galpy.potential import KuzminDiskPotential,MiyamotoNagaiPotential,DoubleExponentialDiskPotential,EllipsoidalPotential,DehnenBarPotential,vcirc,epifreq,dvcircdR
```

```python
KDP = KuzminDiskPotential(a=1,normalize=1)
MNP = MiyamotoNagaiPotential(amp=10,a=0.5,b=0.0375,normalize=1)
EDP = DoubleExponentialDiskPotential(amp=10)
#EP = EllipsoidalPotential()
#DBP = DehnenBarPotential(phi=90)
KDP.plotRotcurve(label='KuzminDiskPotential')
plotRotcurve(MWP,label='Potential of Milkyway',overplot=True)
MNP.plotRotcurve(label='MiyamotoNagaiPotential',overplot=True)
EDP.plotRotcurve(label='DoubleExponentialDiskPotential',overplot=True)
#DBP.plotRotcurve(label='EllipsoidalPotential',overplot=True)
plt.legend()
plt.title('Potentials')
plt.show()
```
