```python
import numpy as np
from einsteinpy.geodesic import Geodesic , Timelike
from einsteinpy . plotting import GeodesicPlotter
```

```python
position = [0.1,4 , 1.5707963310]
# Initial position in Schwarzschildcoordinates
momentum = [-5.05720402,0.4,-1.5]
# Initial momentum
a = 0 # zero for Schwarzschild Black Hole and non - zero for Kerr
# Calculate geodesic
geod = Timelike (metric="Schwarzschild",metric_params=(a,),position=position,momentum=momentum,steps=400,delta=0.5,return_cartesian=True)
print("Geodesic Object:",geod)
gpl=GeodesicPlotter()
gpl.plot(geod,color="green")
gpl.show()
```

```python
gpl.clear() # Clear previous plot
gpl.plot2D( geod , coordinates = (1 , 2 ) , color = " green " )
gpl.show()
```

```python
gpl.clear() # Clear previous plot
gpl.plot2D( geod , coordinates = (1 , 3 ) , color = " green " )
gpl.show()
```

```python
gpl.clear() # Clear previous plot
gpl . parametric_plot ( geod , colors = ( " red " , " green " , " blue " ) )
gpl . show ()
```
