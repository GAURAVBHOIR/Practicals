```python
from numpy import *
import matplotlib.pyplot as plt
import sympy
from einsteinpy.symbolic import RicciTensor, RicciScalar
from einsteinpy.symbolic.predefined import AntiDeSitter,DeSitter,AntiDeSitterStatic

sympy.init_printing()
```

```python
metric = AntiDeSitter()
metric.tensor()
#matric.tensor()
```

```python
Ric = RicciTensor.from_metric(metric)
Ric.tensor()

```

```python
R = RicciScalar.from_riccitensor(Ric)
R.simplify()
R.expr
```

```python
n,nsteps = 20,1000
M=1
r = linspace(1,n,nsteps)
L = [20,40,60]
def ADT(r,L):
    return 1-(2*M/r)+(r**2/L**2)
def DT(r,L):
    return 1-(2*M/r)-(r**2/L**2)
def SM(r,L):
    return 1-(2*M/r)
plt.figure(figsize=(10,10))
for i in L:
    AD = ADT(r,i)
    D =DT(r,i)
    S = SM(r,L)
    plt.plot(r,AD,label=f'AntiDessiter')
    plt.plot(r,D,label=f'Dessiter')
plt.plot(r,S,label='Swarschild')
plt.legend(loc='best')
plt.title('Behaviour of \nAndtiDessiter,Dessiter and Swarschild Metric ')
plt.show()
```
